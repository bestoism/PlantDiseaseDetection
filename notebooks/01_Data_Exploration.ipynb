```python
# Import library yang dibutuhkan
import os
import matplotlib.pyplot as plt
import cv2 
import random

# Tentukan path ke direktori dataset kita
dataset_path = '../data/plantvillage/'

# Dapatkan daftar semua kelas (nama folder) penyakit
classes = os.listdir(dataset_path)
print(f"Total kelas penyakit: {len(classes)}")
print(f"Nama-nama kelas (5 pertama): {classes[:5]}")

# Mari kita coba tampilkan satu gambar secara acak
# 1. Pilih satu kelas acak
random_class_name = random.choice(classes)
class_path = os.path.join(dataset_path, random_class_name)

# 2. Pilih satu gambar acak dari kelas tersebut
random_image_name = random.choice(os.listdir(class_path))
image_path = os.path.join(class_path, random_image_name)

# 3. Baca dan tampilkan gambar
# OpenCV membaca gambar dalam format BGR, kita perlu konversi ke RGB untuk Matplotlib
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8, 8))
plt.imshow(image_rgb)
plt.title(f"Kelas: {random_class_name}\nFile: {random_image_name}")
plt.axis('off') # Sembunyikan sumbu x dan y
plt.show()
```
